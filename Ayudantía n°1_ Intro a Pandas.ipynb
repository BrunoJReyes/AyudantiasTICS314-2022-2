## Pandas

Seleccionar columnas

```python
#Seleccion respecto al nombre y edad de la persona

df_persona = df_estonia[['Firstname',	'Lastname', 'Age']]
df_persona
```

```python
#Condicionales

#Con numeros

df_estonia[df_estonia["Age"] > 50]  #Se aplica la condicional "mayor a 50" a la columna de edad (Age)
```

```python
"""df.groupby(df['columna_para_agrupar'])['columna_operar'].count()"""

# .count() => Contar caracteres 
# .sum() => Sumar numeros (enteros y decimales)
# .mean() => Promedio
```

```python
#Seleccionamos todas las filas donde hayan mujeres

df_muj = df_estonia[df_estonia["Sex"] == "F"]
df_muj = df_muj[["Country", "Sex"]]
df_muj
```

```python
#Agrupemos respecto a las nacionalidades de las personas

df_group = df_muj.groupby(df_muj['Country'])['Sex'].count()
df_group = pd.DataFrame(df_group) #Para correguir los efectos de esta tabla, podemos buscar ciertos comandos en internet que nos ayuden a correguir aquello
df_group
```

```python
df_group = df_group.reset_index() #Aqui finalmente queda reindexado
df_group
```
